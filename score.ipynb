```python
import glob
import numpy as np
import os
```

```python
agents = []
for path in glob.glob("./agent/*.py"):
    file_name = path.split("/")[2]
    agent_name = file_name.split(".")[0]
    print(agent_name)
    if agent_name != "base_agent":
        agents.append(agent_name)
```

```python
def pk(agent1, agent2):
    read_fd, write_fd = os.pipe() 
    processid = os.fork()

    if(processid):
        os.close(write_fd)
        r = os.fdopen(read_fd)  
        read_string = r.read()
        win_rate = read_string.split()[-1]
        status = os.wait() 
        print(f"{agent1} vs {agent2}: {win_rate}")
        return win_rate

        
    else:
        os.dup2(write_fd, 1)
        os.close(read_fd)
        os.execlp("python3", "python3", "arena.py", "--agent1", agent1+".MyAgent", '--agent2', agent2+".MyAgent", "--headless")
```

```python
import multiprocessing
args = []
pool = multiprocessing.Pool(50)
for agent1 in agents:
    for agent2 in agents:
        if agent1 != agent2:
            args.append((agent1, agent2))

ans_list = pool.starmap(pk, args)  
#print(ans_list)
```
